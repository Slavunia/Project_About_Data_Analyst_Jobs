```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

# Lade die DatensÃ¤tze fÃ¼r Junior- und Nicht-Junior-Rollen weltweit
junior_df = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\7_junior_postings_global.csv')
non_junior_df = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\7_nonjunior_postings_global.csv')

# Funktion, um AusreiÃŸer basierend auf der IQR-Methode zu entfernen
def remove_outliers(df, role_type):
    # Berechne Q1 (25. Perzentil) und Q3 (75. Perzentil)
    Q1 = df['salary_year_avg'].quantile(0.25)
    Q3 = df['salary_year_avg'].quantile(0.75)
    
    # Berechne IQR (Interquartilsabstand)
    IQR = Q3 - Q1
    
    # Setze den Multiplikator basierend auf dem Rollentyp
    if role_type == 'junior':
        multiplier = 1.5  # Standardmultiplikator fÃ¼r Junior-Rollen
    else:
        multiplier = 2.5  # Angepasster Multiplikator fÃ¼r Nicht-Junior-Rollen
    
    # Definiere die oberen und unteren Grenzen
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    
    # Filtere den DataFrame, um AusreiÃŸer zu entfernen
    filtered_df = df[(df['salary_year_avg'] >= lower_bound) & (df['salary_year_avg'] <= upper_bound)]
    
    return filtered_df

# Entferne AusreiÃŸer fÃ¼r Junior-Rollen (mit 1.5 Multiplikator)
junior_df_cleaned = remove_outliers(junior_df, 'junior')

# Entferne AusreiÃŸer fÃ¼r Nicht-Junior-Rollen (mit 2.5 Multiplikator)
non_junior_df_cleaned = remove_outliers(non_junior_df, 'non-junior')

# Behebe SettingWithCopyWarning, indem .loc verwendet wird, um sicherzustellen, dass der DataFrame richtig geÃ¤ndert wird
junior_df_cleaned.loc[:, 'job_type'] = 'Junior'
non_junior_df_cleaned.loc[:, 'job_type'] = 'Non-Junior'

# Erstelle zwei separate Histogramme fÃ¼r Junior- und Nicht-Junior-Rollen
plt.figure(figsize=(14, 10))

# Histogramm fÃ¼r Junior-Rollen
plt.subplot(1, 2, 1)  # 1 Zeile, 2 Spalten, 1. Unterdiagramm
sns.histplot(junior_df_cleaned['salary_year_avg'], kde=True, color='blue', bins=20)
plt.title('Gehaltsverteilung fÃ¼r Junior-Rollen (Global)', fontsize=16, fontweight="bold")
plt.xlabel('Durchschnittliches Jahresgehalt (USD)', fontsize=12, fontweight="bold")
plt.ylabel('HÃ¤ufigkeit', fontsize=12, fontweight="bold")
plt.grid(True)  # Gitterlinien hinzufÃ¼gen

# Histogramm fÃ¼r Nicht-Junior-Rollen
plt.subplot(1, 2, 2)  # 1 Zeile, 2 Spalten, 2. Unterdiagramm
sns.histplot(non_junior_df_cleaned['salary_year_avg'], kde=True, color='green', bins=20)
plt.title('Gehaltsverteilung fÃ¼r Nicht-Junior-Rollen (Global)', fontsize=16, fontweight="bold")
plt.xlabel('Durchschnittliches Jahresgehalt (USD)', fontsize=12, fontweight="bold")
plt.ylabel('HÃ¤ufigkeit', fontsize=12, fontweight="bold")
plt.grid(True)  # Gitterlinien hinzufÃ¼gen

plt.tight_layout()

# Definiere den Pfad zum "python_visualizations"-Ordner
visualizations_folder = 'python_visualizations'

# Stelle sicher, dass der Ordner existiert
if not os.path.exists(visualizations_folder):
    os.makedirs(visualizations_folder)

# Speichere das Diagramm direkt im "python_visualizations"-Ordner
plt.savefig(os.path.join(visualizations_folder, '07_gehaltsverteilung_bereinigt.png'), dpi=300, bbox_inches='tight', facecolor='white')

# Zeige das Diagramm an
plt.show()
```

```python
import os 
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import re

# Lade den Datensatz fÃ¼r den globalen Markt
df_global = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\08_employment_type_global.csv')

# Entferne Zeilen, bei denen 'job_schedule_type' NaN ist
df_global = df_global.dropna(subset=['job_schedule_type'])

# Schritt 1: Entferne "Pekerjaan tetap" Zeilen
df_global = df_global[df_global['job_schedule_type'] != 'Pekerjaan tetap']

# Schritt 2: Definiere mehrteilige AusdrÃ¼cke, die durch Platzhalter ersetzt werden sollen
multi_word_phrases = ['Temp work', 'Per diem']

# Schritt 3: Ersetze mehrteilige AusdrÃ¼cke durch Platzhalter, um eine Aufspaltung zu vermeiden
df_global['job_schedule_type'] = df_global['job_schedule_type'].replace({
    'Temp work': 'TEMP_WORK',
    'Per diem': 'PER_DIEM'
}, regex=False)

# Schritt 4: Teile BeschÃ¤ftigungsarten durch 'and', ', ' und ', and' und zÃ¤hle einzelne Komponenten
individual_components_global = []

# Iteriere Ã¼ber die BeschÃ¤ftigungsarten und teile sie durch die angegebenen Trennzeichen
for schedule in df_global['job_schedule_type'].dropna():
    components_global = re.split(r'\s*(?:, and|and|,)\s*', schedule)  # Teile nach 'and', ', and' oder ', '
    individual_components_global.extend(components_global)

# Schritt 5: ZÃ¤hle die Vorkommen jeder Komponente
component_counts_global = Counter(individual_components_global)

# Schritt 6: Ersetze Platzhalter zurÃ¼ck mit den korrekten mehrteiligen AusdrÃ¼cken
component_counts_global = {key.replace('TEMP_WORK', 'Zeitarbeit').replace('PER_DIEM', 'Tagessatz'): value 
                           for key, value in component_counts_global.items()}

# Schritt 7: Berechne die Gesamtfrequenz fÃ¼r den globalen Markt
total_frequency_global = sum(component_counts_global.values())

# Schritt 8: Gruppiere BeschÃ¤ftigungsarten, die weniger als 5% ausmachen, in "Sonstige" fÃ¼r den globalen Markt
grouped_counts_global = {}
others_count_global = 0
others_details_global = []

# Ãœbersetzungen WÃ¶rterbuch der BeschÃ¤ftigungsarten 
job_schedule_translations = {
    'Full-time': 'Vollzeit',
    'Contractor': 'SelbststÃ¤ndig', 
    'Part-time': 'Teilzeit',
    'Internship': 'Praktikum',
    'Temp work': 'Zeitarbeit',
    'Volunteer': 'Ehrenamt',  
    'Per diem': 'Tagessatz',
    'Others': 'Sonstige'
}

# Schritt 9: Ersetze alle BeschÃ¤ftigungsarten im DataFrame und in den ZÃ¤hlungen
component_counts_global_translated = {
    job_schedule_translations.get(key, key): value
    for key, value in component_counts_global.items()
}

# Schritt 10: Gruppiere die BeschÃ¤ftigungsarten mit weniger als 5% in "Sonstige"
grouped_counts_global = {}
others_count_global = 0
others_details_global = []

# Identifiziere Komponenten, die in "Sonstige" gruppiert werden sollten
for job_type, count in component_counts_global_translated.items():
    percentage = (count / total_frequency_global) * 100
    if percentage < 5:
        others_count_global += count
        others_details_global.append((job_type, count, percentage))  # Speichere als Tuple mit Prozentwert zur Sortierung
    else:
        grouped_counts_global[job_type] = count

# Sortiere die "Sonstige"-Komponenten nach HÃ¤ufigkeit in absteigender Reihenfolge
others_details_global.sort(key=lambda x: x[1], reverse=True)  # Sortiere nach HÃ¤ufigkeit (Index 1)

# FÃ¼ge die "Sonstige"-Kategorie zu den gruppierten ZÃ¤hlungen hinzu
if others_count_global > 0:
    grouped_counts_global[job_schedule_translations['Others']] = others_count_global

# Schritt 11: Erstelle einen DataFrame fÃ¼r eine bessere Visualisierung fÃ¼r den globalen Markt
component_df_global = pd.DataFrame(component_counts_global_translated.items(), columns=['BeschÃ¤ftigungsart', 'HÃ¤ufigkeit'])
component_df_global = component_df_global.sort_values(by='HÃ¤ufigkeit', ascending=False)

updated_component_df_global = pd.DataFrame(grouped_counts_global.items(), columns=['BeschÃ¤ftigungsart', 'HÃ¤ufigkeit'])
updated_component_df_global = updated_component_df_global.sort_values(by='HÃ¤ufigkeit', ascending=False)

# Berechne den Prozentsatz fÃ¼r jede BeschÃ¤ftigungsart in der aktualisierten Tabelle fÃ¼r den globalen Markt
component_df_global['Prozent'] = (component_df_global['HÃ¤ufigkeit'] / component_df_global['HÃ¤ufigkeit'].sum()) * 100
component_df_global['Prozent'] = component_df_global['Prozent'].round(2)  # Stelle sicher, dass es 2 Dezimalstellen gibt

updated_component_df_global['Prozent'] = (updated_component_df_global['HÃ¤ufigkeit'] / updated_component_df_global['HÃ¤ufigkeit'].sum()) * 100
updated_component_df_global['Prozent'] = updated_component_df_global['Prozent'].round(2)  # Stelle sicher, dass es 2 Dezimalstellen gibt

# Zeige das Ergebnis als Tabelle fÃ¼r den globalen Markt
print("HÃ¤ufigkeit der BeschÃ¤ftigungsarten (Global)")
print(component_df_global)

# Zeige das Ergebnis als Tabelle fÃ¼r den globalen Markt
print("Aktualisierte HÃ¤ufigkeit der BeschÃ¤ftigungsarten (Global)")
print(updated_component_df_global)

# Schritt 12: Plotte die Ergebnisse als Kreisdiagramm fÃ¼r den globalen Markt
plt.figure(figsize=(10, 10))

# Kreisdiagramm fÃ¼r die gruppierten Daten (globaler Markt)
plt.pie(updated_component_df_global['HÃ¤ufigkeit'], labels=updated_component_df_global['BeschÃ¤ftigungsart'], autopct='%1.1f%%', startangle=140, 
        colors=plt.cm.Paired.colors, wedgeprops={'edgecolor': 'black'})

# FÃ¼ge einen Titel fÃ¼r den globalen Markt hinzu
plt.title('Verteilung der BeschÃ¤ftigungsarten in Datenanalysten-Rollen (Global)', fontsize=18, fontweight="bold")

# FÃ¼ge eine dynamische Legende fÃ¼r die "Sonstige"-Kategorie mit detaillierter AufschlÃ¼sselung fÃ¼r den globalen Markt hinzu
others_legend_global = "\n".join([f'- {job_type}: {count} ({percentage:.2f}%)' for job_type, count, percentage in others_details_global])
labels_global = [
    f'{key}: {value} ({value/total_frequency_global*100:.2f}%)' 
    for key, value in component_counts_global_translated.items() if (value / total_frequency_global) * 100 >= 5
] + [f'Sonstige: {others_count_global} ({others_count_global/total_frequency_global*100:.2f}%)\n{others_legend_global}']

# Erstelle die Legende mit verbesserter Lesbarkeit fÃ¼r "Sonstige" fÃ¼r den globalen Markt
plt.legend(labels_global, loc='center left', bbox_to_anchor=(1, 0.5), title="BeschÃ¤ftigungsarten", fontsize=12, handlelength=3, handleheight=1)

# Zeige das Kreisdiagramm an
plt.axis('equal')  # Gleiche Aspect-Ratio stellt sicher, dass das Diagramm als Kreis gezeichnet wird.
plt.tight_layout()

# Definiere den Pfad zum "python_visualizations"-Ordner
visualizations_folder = 'python_visualizations'

# Stelle sicher, dass der Ordner existiert
if not os.path.exists(visualizations_folder):
    os.makedirs(visualizations_folder)

# Speichere das Diagramm direkt im "python_visualizations"-Ordner
plt.savefig(os.path.join(visualizations_folder, '08_beschaeftigungsarten_global.png'), dpi=300, bbox_inches='tight', facecolor='white')

# Zeige das Diagramm an
plt.show()
```

```python
import os
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import re

# Lade den Datensatz fÃ¼r Deutschland
df_germany = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\08_employment_type_germany.csv')

# Entferne Zeilen, bei denen 'job_schedule_type' NaN ist
df_germany = df_germany.dropna(subset=['job_schedule_type'])

# Schritt 1: Entferne "Pekerjaan tetap" Zeilen
df_germany = df_germany[df_germany['job_schedule_type'] != 'Pekerjaan tetap']

# Schritt 2: Definiere mehrteilige AusdrÃ¼cke, die durch Platzhalter ersetzt werden sollen
multi_word_phrases = ['Temp work', 'Per diem']

# Schritt 3: Ersetze mehrteilige AusdrÃ¼cke durch Platzhalter, um eine Aufspaltung zu vermeiden
df_germany['job_schedule_type'] = df_germany['job_schedule_type'].replace({
    'Temp work': 'TEMP_WORK',
    'Per diem': 'PER_DIEM'
}, regex=False)

# Schritt 4: Teile BeschÃ¤ftigungsarten durch 'and', ', ' und ', and' und zÃ¤hle einzelne Komponenten
individual_components_germany = []

# Iteriere Ã¼ber die BeschÃ¤ftigungsarten und teile sie durch die angegebenen Trennzeichen
for schedule in df_germany['job_schedule_type'].dropna():
    components = re.split(r'\s*(?:, and|and|,)\s*', schedule)  # Teile nach 'and', ', and' oder ', '
    individual_components_germany.extend(components)

# Schritt 5: ZÃ¤hle die Vorkommen jeder Komponente
component_counts_germany = Counter(individual_components_germany)

# Schritt 6: Ersetze Platzhalter zurÃ¼ck mit den korrekten mehrteiligen AusdrÃ¼cken
component_counts_germany = {key.replace('TEMP_WORK', 'Zeitarbeit').replace('PER_DIEM', 'Tagessatz'): value 
                            for key, value in component_counts_germany.items()}

# Schritt 7: Berechne die Gesamtfrequenz
total_frequency_germany = sum(component_counts_germany.values())

# Schritt 8: Gruppiere BeschÃ¤ftigungsarten, die weniger als 5% ausmachen, in "Sonstige"
grouped_counts_germany = {}
others_count_germany = 0
others_details_germany = []

# Ãœbersetzungen WÃ¶rterbuch der BeschÃ¤ftigungsarten
job_schedule_translations = {
    'Full-time': 'Vollzeit',
    'Contractor': 'SelbststÃ¤ndig',  
    'Part-time': 'Teilzeit',
    'Internship': 'Praktikum',
    'Temp work': 'Zeitarbeit',
    'Volunteer': 'Ehrenamt',  
    'Per diem': 'Tagessatz',
    'Others': 'Sonstige'
}

# Schritt 9: Ersetze alle BeschÃ¤ftigungsarten im DataFrame und in den ZÃ¤hlungen
component_counts_germany_translated = {
    job_schedule_translations.get(key, key): value
    for key, value in component_counts_germany.items()
}

# Schritt 10: Gruppiere die BeschÃ¤ftigungsarten mit weniger als 5% in "Sonstige"
grouped_counts_germany = {}
others_count_germany = 0
others_details_germany = []

# Identifiziere Komponenten, die in "Sonstige" gruppiert werden sollten
for job_type, count in component_counts_germany_translated.items():
    percentage = (count / total_frequency_germany) * 100
    if percentage < 5:
        others_count_germany += count
        others_details_germany.append((job_type, count, percentage))  # Speichere als Tuple mit Prozentwert zur Sortierung
    else:
        grouped_counts_germany[job_type] = count

# Sortiere die "Sonstige"-Komponenten nach HÃ¤ufigkeit in absteigender Reihenfolge
others_details_germany.sort(key=lambda x: x[1], reverse=True)  # Sortiere nach HÃ¤ufigkeit (Index 1)

# FÃ¼ge die "Sonstige"-Kategorie zu den gruppierten ZÃ¤hlungen hinzu
if others_count_germany > 0:
    grouped_counts_germany[job_schedule_translations['Others']] = others_count_germany

# Schritt 11: Erstelle einen DataFrame fÃ¼r eine bessere Visualisierung fÃ¼r Deutschland
component_df_germany = pd.DataFrame(component_counts_germany_translated.items(), columns=['BeschÃ¤ftigungsart', 'HÃ¤ufigkeit'])
component_df_germany = component_df_germany.sort_values(by='HÃ¤ufigkeit', ascending=False)

updated_component_df_germany = pd.DataFrame(grouped_counts_germany.items(), columns=['BeschÃ¤ftigungsart', 'HÃ¤ufigkeit'])
updated_component_df_germany = updated_component_df_germany.sort_values(by='HÃ¤ufigkeit', ascending=False)

# Berechne den Prozentsatz fÃ¼r jede BeschÃ¤ftigungsart in der Tabelle
component_df_germany['Prozent'] = (component_df_germany['HÃ¤ufigkeit'] / component_df_germany['HÃ¤ufigkeit'].sum()) * 100
component_df_germany['Prozent'] = component_df_germany['Prozent'].round(2)  # Stelle sicher, dass es 2 Dezimalstellen gibt

# Berechne den Prozentsatz fÃ¼r jede BeschÃ¤ftigungsart in der aktualisierten Tabelle
updated_component_df_germany['Prozent'] = (updated_component_df_germany['HÃ¤ufigkeit'] / updated_component_df_germany['HÃ¤ufigkeit'].sum()) * 100
updated_component_df_germany['Prozent'] = updated_component_df_germany['Prozent'].round(2)  # Stelle sicher, dass es 2 Dezimalstellen gibt

# Zeige das Ergebnis als Tabelle fÃ¼r Deutschland
print("HÃ¤ufigkeit der BeschÃ¤ftigungsarten (Deutschland)")
print(component_df_germany)

# Zeige das aktualisierte Ergebnis als Tabelle fÃ¼r Deutschland
print("Aktualisierte HÃ¤ufigkeit der BeschÃ¤ftigungsarten (Deutschland)")
print(updated_component_df_germany)

# Schritt 12: Plotte die Ergebnisse als Kreisdiagramm fÃ¼r Deutschland
plt.figure(figsize=(10, 10))

# Kreisdiagramm fÃ¼r die gruppierten Daten (Deutschland)
plt.pie(updated_component_df_germany['HÃ¤ufigkeit'], labels=updated_component_df_germany['BeschÃ¤ftigungsart'], autopct='%1.1f%%', startangle=140, 
        colors=plt.cm.Paired.colors, wedgeprops={'edgecolor': 'black'})

# FÃ¼ge einen Titel hinzu
plt.title('Verteilung der BeschÃ¤ftigungsarten in Datenanalysten-Rollen (Deutschland)', fontsize=18, fontweight="bold")

# FÃ¼ge eine dynamische Legende fÃ¼r die "Sonstige"-Kategorie mit detaillierter AufschlÃ¼sselung, sortiert nach HÃ¤ufigkeit, hinzu
others_legend_germany = "\n".join([f'- {job_type}: {count} ({percentage:.2f}%)' for job_type, count, percentage in others_details_germany])
labels_germany = [
    f'{key}: {value} ({value/total_frequency_germany*100:.2f}%)' 
    for key, value in component_counts_germany_translated.items() if (value / total_frequency_germany) * 100 >= 5
] + [f'Sonstige: {others_count_germany} ({others_count_germany/total_frequency_germany*100:.2f}%)\n{others_legend_germany}']

# Erstelle die Legende mit verbesserter Lesbarkeit fÃ¼r "Sonstige"
plt.legend(labels_germany, loc='center left', bbox_to_anchor=(1, 0.5), title="BeschÃ¤ftigungsarten", fontsize=12, handlelength=3, handleheight=1)

# Zeige das Kreisdiagramm an
plt.axis('equal')  # Gleiche Aspect-Ratio stellt sicher, dass das Diagramm als Kreis gezeichnet wird.
plt.tight_layout()

# Definiere den Pfad zum "python_visualizations"-Ordner
visualizations_folder = 'python_visualizations'

# Stelle sicher, dass der Ordner existiert
if not os.path.exists(visualizations_folder):
    os.makedirs(visualizations_folder)

# Speichere das Diagramm direkt im "python_visualizations"-Ordner
plt.savefig(os.path.join(visualizations_folder, '08_beschaeftigungsarten_deutschland.png'), dpi=300, bbox_inches='tight', facecolor='white')

# Zeige das Diagramm an
plt.show()
```
